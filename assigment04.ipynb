```python
import math

class Circle:
    def __init__(self, r):
        self.r = r
    def area(self):
        return math.pi * self.r * self.r
    def perimeter(self):
        return 2 * math.pi * self.r

c = Circle(5)
print(c.area())
print(c.perimeter())
```

```python
from datetime import date

class Person:
    def __init__(self, name, country, dob):
        self.name = name
        self.country = country
        self.dob = dob  # YYYY-MM-DD

    def age(self):
        y, m, d = map(int, self.dob.split('-'))
        b = date(y, m, d)
        t = date.today()
        return t.year - b.year - ((t.month, t.day) < (b.month, b.day))

p = Person("Hadi", "Pakistan", "1985-08-12")
print(p.age())
```

```python
class Calculator:
    def add(self, a, b): return a + b
    def subtract(self, a, b): return a - b
    def multiply(self, a, b): return a * b
    def divide(self, a, b): return a / b if b != 0 else "Cannot divide by zero"

c = Calculator()
print(c.add(10, 5))
```

```python
numbers = [12, 45, 2, 89, 33]

largest = numbers[0]
smallest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num
    if num < smallest:
        smallest = num

print(f"Largest: {largest}")
print(f"Smallest: {smallest}")
```

```python
numbers = [12, 5, 8, 20, 15, 3, 25, 10, 18, 4]

total = len(numbers)
total_sum = sum(numbers)
average = total_sum / total
max_val = max(numbers)
min_val = min(numbers)

print(f"Total numbers: {total}")
print(f"Sum: {total_sum}")
print(f"Average: {average}")
print(f"Maximum: {max_val}")
print(f"Minimum: {min_val}")
```
