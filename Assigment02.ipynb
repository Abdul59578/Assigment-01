```python
n = int(input("Enter a number: "))

print(f"The first {n} terms of the series are:")
for i in range(1, n + 1):
    print(i ** 2, end=" ")
```

```python
num = int(input("Enter a number: "))

print(f"Multiplication table of {num}:")
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

```python
n = int(input("Enter a number: "))

print(f"Odd numbers between 1 and {n}:")
for i in range(1, n + 1):
    if i % 2 != 0:
        print(i, end=" ")
```

```python
num = 20
while num >= 1:
    print(num)
    num -= 1
```

```python
n = int(input("Enter a number: "))

print(f"Numbers from 1 to {n}:")
for i in range(1, n + 1):
    print(i)
```

```python
total = 10

for i in range(5):
    num = int(input("Enter a number: "))
    total += num

print(f"The final running total is: {total}")
```

```python
clean_countries = ["Bangladesh", "India", "Pakistan", "China"]

while True:
    country = input("Enter a country name: ")
    if country in clean_countries:
        print(f"{country} is in clean countries list")
        break
```
