```python
# Bio using variables
name = "Abdul Hadi"
father_name = "Bashir"
dob = "4/1/2014"
age = 11
hobby = "reading and drawing"
city = "Karachi"


print(" My Bio ")
print("Name:\t\t", name)
print("Father Name:\t", father_name)
print("DOB:\t\t", dob)
print("Age:\t\t", age, "years old")
print("Hobby:\t\t", hobby)
print("City:\t\t", city)
```

```python
a = 10
b = 5

print(a + b)   # add = 15
print(a - b)   # subtract = 5
print(a * b)   # multiply = 50
print(a / b)   # divide = 2.0

print(a > b)   # bigger than? True
print(a == b)  # same? False

a = a + 1      # make a bigger by 1
print(a)       # 11
```
